```python
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup

import csv
import pandas as pd
```

# Scraping PreFlop Poker Hands Rank

## 1. Scraping:

```python
def getHtml(news_url, variation):
    if variation == 1:
        html_page = urlopen(news_url)
        return BeautifulSoup(html_page, 'lxml')
    if variation == 2:
        hdr = {'User-Agent': 'Mozilla/5.0'} 
        req = Request(news_url,headers=hdr)
        html_page = urlopen(req)
        return BeautifulSoup(html_page, 'lxml')
```

```python
url = "https://www.preflophands.com/"
soup_html = getHtml(url,2)

cards = soup_html.findAll("p")
cards = cards[9:]
```

## 2. Treatig information:

```python
raw_cards = []
cnt = 0
for texts in cards:
    cnt += 1
    
    if cnt == 3:
        raw_cards.append("#3QQ")
    
    elif cnt == 52:
        raw_cards.append("#5222")
        
    elif cnt == 74:
        raw_cards.append("#74T6s")
        
    else:
        try:
            raw_cards.append(texts.font.text)
        except:
            raw_cards.append(texts.span.text)
            
#Treating a outlier:
raw_cards[50] = raw_cards[50][:-3]
```

```python
treated_cards = []

for cnt in range(len(raw_cards)):
    
    
    if (cnt+1) < 10:
        num = 2
    elif (cnt+1) < 100:
        num = 3
    else:
        num = 4
               
    word = raw_cards[cnt]
    
    rank = word[1:num]
            
    if  word[-1].isdigit() or word[-1] == "T" or word[-1] == "J" or word[-1] == "Q" or word[-1] == "K" or word[-1] == "A":
        treated_cards.append([rank, word[num:], 'p'])
    else:
        treated_cards.append([rank, word[num:-1], word[-1]])
        
    
```

```python
for a in range(len(treated_cards)):
    print(treated_cards[a])
```

## 3. Import into CSV

```python
df = pd.DataFrame(treated_cards) # construct data frame and transpose

df.to_csv('preFlop-rank.csv', header=False, index=False, sep=',')
```
